# Saperate Predictors and Target

```python
X = df.iloc[:,:-1]
y = df.outcome
```

# Saperate the Predictors and Target into training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=35)
```

# Training

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2500)
```

```python
model.fit(X_train, y_train)
```

# Evaluate

```python
y_predict = model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict)
```

# Assignment Solution

```python
import seaborn as sns
import numpy as np
from matplotlib import rcParams
```

```python
rcParams['figure.figsize'] = (10,10)
```

```python
sns.heatmap(X.corr(), annot=True)
```

```python
X.corr()
```

```python
def getNextBigCorr(df, thresh=0.5):
    corr = X.corr()
    corr.values[np.triu_indices_from(corr, 0)] = np.nan
    for i in corr.columns:
        for j in corr.index:
            if np.abs(corr.loc[j,i]) > thresh and i != j:
                return j
    return 0
```

```python
X = df.iloc[:,:-1]
while getNextBigCorr(X) != 0:
    col = getNextBigCorr(X)
    X.drop([col], axis=1, inplace=True)
```
